### Create dictionary of amino acid one-letter codes

```python
import os
import pandas as pd
from tqdm import tqdm
import pickle
```

```python
one_letter_code = {'3 code': ["ALA", "ARG", "ASN", "ASP", "CYS", "GLN", "GLU", "GLY", "HIS",
                            "ILE", "LEU", "LYS", "MET", "PHE", "PRO", "SER", "THR", "TRP", "TYR", "VAL"],
                  '1 code': ["A", "R", "N", "D", "C", "Q", "E", "G", "H", "I", "L", "K", "M", "F",
                             "P", "S", "T", "W", "Y", "V"]}
df_encode = pd.DataFrame.from_dict(one_letter_code)
df_encode.head()
```

```python
mother_of_all_sources = '/media/the_beast/disk/protein_folding/amino_acid_tags/'
```

```python
dictionary = {}
dictionary.clear()
for root, dirs, files in os.walk(mother_of_all_sources, topdown=False):
    for name in files:
        try:
            df_amino = pd.read_csv(mother_of_all_sources + name, header=None)
        except:
            pass
        df_amino.rename(columns={0:'3 code'}, inplace=True)
        df_merge = pd.merge(df_amino, df_encode, on='3 code', how='left')
        df_merge.drop(columns=['3 code'], axis=1, inplace=True)
        df_list = df_merge['1 code'].values.tolist()
        try:
            new_list = ''.join(df_list)
        except:
            pass
        dictionary.update({name.split('_')[0]:new_list})
print(len(dictionary))
```

```python
checklist = []
checklist.clear()
unique_items = {}
unique_items.clear()
duplicates = {}
duplicates.clear()
for key, value in dictionary.items():
    if value in checklist:
        duplicates.update({key:value})
    else:
        checklist.append(value)
        unique_items.update({key:value})
print(len(dictionary))
print(len(unique_items))
print(len(duplicates))
```

```python
unique_values = []
unique_values.clear()
for key, value in unique_items.items():
    unique_values.append(key)
print(len(unique_values))
```
