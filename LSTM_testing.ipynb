### Check same amino acids are in all three datatypes (adjacency matrix, features, labels)

```python
# source_a = '/media/the_beast/A/mathisi_tests/data/nearest_neighbour_8_normalised/'
tag_list = []
tag_list.clear()
for root, dirs, files in os.walk(source_a, topdown=False):
    for name in files:
        tag_list.append(name.split('_')[0])
print(tag_list)
print(len(tag_list))
```

```python
tag_counter = {}
for tag in tag_list:
    tag_count = tag_list.count(tag)
    tag_counter.update({tag:tag_count})
```

```python
# print(tag_counter)
```

```python
key_list = []
key_list.clear()
for key, value in tag_counter.items():
    if value != 3:
        key_list.append(key)
print(key_list)
```

```python
file_path = '/media/the_beast/A/mathisi_tests/data/sample_100/test_data/'
for root, dirs, files in os.walk(source_a, topdown=False):
    for name in files:
        if name.split('_')[0] in key_list:
            try:
                os.remove(file_path + name.split('_')[0] + '_adjacency-matrix.pickle')
            except:
                pass
            try:
                os.remove(file_path + name.split('_')[0] + '_features.pickle')
            except:
                pass
            try:
                os.remove(file_path + name.split('_')[0] + '_labels.pickle')
            except:
                pass
```

### Get max distance value from all distance matricies 

```python
import os
import torch
import tqdm
```

```python
source = '/media/the_beast/A/mathisi_tests/data/distance_matrix/'
```

```python
max_values = []
max_values.clear()
for root, dirs, files in os.walk(source, topdown=False):
    print(len(files))
    for file in files:
        with open(source + file, 'rb') as labels_file:
            try:
                tensor = pickle.load(labels_file)
                max_values.append(torch.max(tensor))
            except:
                pass
print(max_values)
print(len(max_values))
```
